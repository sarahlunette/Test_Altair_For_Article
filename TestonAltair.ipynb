```python
baby_names = baby_names[baby_names['Count'] > 20000]
```

```python
(
    alt.Chart(baby_names)
    .encode(
        alt.X("Year:T", title="Année"),
        alt.Y("Count:Q", axis=alt.Axis(format="~s")),
        alt.Color("Name:N", title="Nom"),
        alt.Order("Name:Q", sort="descending"),
    )
    .mark_area()
    .transform_joinaggregate(mean_count="mean(Count)", groupby=["Name"])
    .transform_filter(alt.FieldOneOfPredicate(field='Name', oneOf=['Amanda', 'Christopher', 'David', 'James', 'Jason', 'Jennifer',
       'John', 'Joshua', 'Matthew', 'Michael']))
    .properties(width=400, height=200)
)
```
